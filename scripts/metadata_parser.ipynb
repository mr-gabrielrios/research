```python
def metadata_parser(data):
    metadata = []
    for data_var in data.data_vars:
        print(data_var)
        temp = data[data_var].attrs
        temp['var'] = data_var
        metadata.append(data[data_var].attrs)
    
    df = pd.DataFrame(metadata)
    cols = list(df.columns.values)
    var_idx = cols.index('var')
    cols = cols[var_idx:var_idx+1] + cols[:var_idx] + cols[var_idx+1:]
    df = df[cols]
    df.to_csv('/projects/GEOCLIM/grios/reference/AM2.5_var_metadata-{0}_{1}.csv'.format(postp_type[0], postp_type[1]), index=False)

metadata_parser(data)
```
